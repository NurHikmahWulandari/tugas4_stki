# 1. Hashes

Hashes (hash table) adalah struktur data yang menggunakan fungsi hash untuk mengonversi kunci menjadi indeks dalam array, memungkinkan pencarian, penambahan, dan penghapusan dalam waktu konstan dalam kasus rata-rata

```python
class HashTableBarang:
    def __init__(self):
        self.hash_table = {}

    def tambah_barang(self, nama, id_barang, harga):
        if id_barang not in self.hash_table:
            self.hash_table[id_barang] = {'nama': nama, 'harga': harga}
            print(f"Barang {nama} berhasil ditambahkan.")
        else:
            print(f"Barang dengan ID {id_barang} sudah ada dalam daftar.")

    def cari_barang(self, id_barang):
        if id_barang in self.hash_table:
            print("Informasi Barang:")
            print(f"Nama: {self.hash_table[id_barang]['nama']}")
            print(f"Harga: {self.hash_table[id_barang]['harga']}")
        else:
            print("Barang tidak ditemukan.")

    def hapus_barang(self, id_barang):
        if id_barang in self.hash_table:
            del self.hash_table[id_barang]
            print(f"Barang dengan ID {id_barang} berhasil dihapus.")
        else:
            print("Barang tidak ditemukan.")

    def daftar_semua_barang(self):
        print("Daftar Barang:")
        for id_barang, info in self.hash_table.items():
            print(f"ID Barang: {id_barang}, Nama: {info['nama']}, Harga: {info['harga']}")

def main():
    daftar_barang = HashTableBarang()

    while True:
        print("\nPilih menu:")
        print("1. Tambah Barang")
        print("2. Cari Barang")
        print("3. Hapus Barang")
        print("4. Daftar Semua Barang")
        print("5. Keluar")

        pilihan = input("Masukkan pilihan (1/2/3/4/5): ")

        if pilihan == '1':
            nama = input("Masukkan nama barang: ")
            id_barang = input("Masukkan ID barang: ")
            harga = input("Masukkan harga barang: ")
            daftar_barang.tambah_barang(nama, id_barang, harga)
        elif pilihan == '2':
            id_barang = input("Masukkan ID barang yang ingin dicari: ")
            daftar_barang.cari_barang(id_barang)
        elif pilihan == '3':
            id_barang = input("Masukkan ID barang yang ingin dihapus: ")
            daftar_barang.hapus_barang(id_barang)
        elif pilihan == '4':
            daftar_barang.daftar_semua_barang()
        elif pilihan == '5':
            print("Terima kasih, Sampai Jumpa!")
            break
        else:
            print("Pilihan tidak valid. Silakan masukkan nomor menu yang benar.")

if __name__ == "__main__":
    main()
```

Code di atas adalah sebuah aplikasi sederhana yang memungkinkan pengguna untuk mengelola daftar barang menggunakan struktur data hash table. Dengan menggunakan hash table, program ini memungkinkan pengguna untuk dengan mudah menambahkan, mencari, menghapus, dan menampilkan daftar barang dengan efisien, karena operasi-operasi tersebut dilakukan dalam waktu konstan (O(1)) dalam kasus rata-rata

Kelas HashTableBarang:
- Kelas ini bertanggung jawab untuk menyimpan dan mengelola daftar barang menggunakan struktur data hash table.
- Pada konstruktor __init__, sebuah hash table kosong dibuat menggunakan dictionary Python.
- Metode tambah_barang digunakan untuk menambahkan barang baru ke dalam hash table. Jika id_barang belum ada dalam hash table, barang baru akan ditambahkan dengan kunci id_barang. Jika barang dengan id_barang yang sama sudah ada, pesan kesalahan akan dicetak.
- Metode cari_barang digunakan untuk mencari barang berdasarkan id_barang dalam hash table. Jika barang ditemukan, informasi tentang barang tersebut (nama dan harga) akan dicetak.
- Metode hapus_barang digunakan untuk menghapus barang dari hash table berdasarkan id_barang. Jika barang ditemukan, barang tersebut akan dihapus dari hash table.
- Metode daftar_semua_barang digunakan untuk mencetak daftar semua barang yang ada dalam hash table, termasuk id_barang, nama, dan harga.

Fungsi main:
- Fungsi main adalah titik masuk utama program.
- Program akan menampilkan menu kepada pengguna dan meminta input untuk memilih operasi yang diinginkan.
- Setiap operasi yang dipilih oleh pengguna (tambah, cari, hapus, tampilkan, atau keluar) akan memanggil metode yang sesuai dari kelas HashTableBarang.
- Program akan terus berjalan hingga pengguna memilih untuk keluar dengan memilih opsi "5".

Input dan Output:
- Program menggunakan fungsi input() untuk menerima input dari pengguna dan print() untuk mencetak output.
- Setiap operasi yang dilakukan oleh pengguna akan memberikan umpan balik ke layar, misalnya, konfirmasi penambahan barang, informasi barang yang ditemukan, pesan kesalahan jika barang tidak ditemukan, dll.

# 2. Trees

Trees (pohon) adalah struktur data hierarkis di mana setiap simpul memiliki dua anak simpul (kiri dan kanan), memungkinkan penyimpanan data secara terurut dan operasi-operasi seperti pencarian, penambahan, dan penghapusan dengan kompleksitas waktu logaritmik dalam kasus rata-rata

```python
class TreeNode:
    def __init__(self, id_barang, nama, harga):
        self.id_barang = id_barang
        self.nama = nama
        self.harga = harga
        self.left = None
        self.right = None

class BinarySearchTreeBarang:
    def __init__(self):
        self.root = None

    def tambah_barang(self, id_barang, nama, harga):
        self.root = self._tambah_recursive(self.root, id_barang, nama, harga)

    def _tambah_recursive(self, node, id_barang, nama, harga):
        if node is None:
            return TreeNode(id_barang, nama, harga)
        if id_barang < node.id_barang:
            node.left = self._tambah_recursive(node.left, id_barang, nama, harga)
        elif id_barang > node.id_barang:
            node.right = self._tambah_recursive(node.right, id_barang, nama, harga)
        else:
            print(f"Barang dengan ID {id_barang} sudah ada dalam daftar.")
        return node

    def cari_barang(self, id_barang):
        return self._cari_recursive(self.root, id_barang)

    def _cari_recursive(self, node, id_barang):
        if node is None or node.id_barang == id_barang:
            return node
        if id_barang < node.id_barang:
            return self._cari_recursive(node.left, id_barang)
        return self._cari_recursive(node.right, id_barang)

    def hapus_barang(self, id_barang):
        self.root = self._hapus_recursive(self.root, id_barang)

    def _hapus_recursive(self, node, id_barang):
        if node is None:
            return node
        if id_barang < node.id_barang:
            node.left = self._hapus_recursive(node.left, id_barang)
        elif id_barang > node.id_barang:
            node.right = self._hapus_recursive(node.right, id_barang)
        else:
            if node.left is None:
                return node.right
            elif node.right is None:
                return node.left
            temp = self._minimum_value_node(node.right)
            node.id_barang = temp.id_barang
            node.nama = temp.nama
            node.harga = temp.harga
            node.right = self._hapus_recursive(node.right, temp.id_barang)
        return node

    def _minimum_value_node(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    def daftar_semua_barang(self):
        self._inorder_traversal(self.root)

    def _inorder_traversal(self, node):
        if node is not None:
            self._inorder_traversal(node.left)
            print(f"ID Barang: {node.id_barang}, Nama: {node.nama}, Harga: {node.harga}")
            self._inorder_traversal(node.right)

def main():
    daftar_barang = BinarySearchTreeBarang()

    while True:
        print("\nPilih menu:")
        print("1. Tambah Barang")
        print("2. Cari Barang")
        print("3. Hapus Barang")
        print("4. Daftar Semua Barang")
        print("5. Keluar")

        pilihan = input("Masukkan pilihan (1/2/3/4/5): ")

        if pilihan == '1':
            id_barang = int(input("Masukkan ID barang: "))
            nama = input("Masukkan nama barang: ")
            harga = int(input("Masukkan harga barang: "))
            daftar_barang.tambah_barang(id_barang, nama, harga)
        elif pilihan == '2':
            id_barang = int(input("Masukkan ID barang yang ingin dicari: "))
            barang = daftar_barang.cari_barang(id_barang)
            if barang:
                print("Informasi Barang:")
                print(f"ID Barang: {barang.id_barang}, Nama: {barang.nama}, Harga: {barang.harga}")
            else:
                print("Barang tidak ditemukan.")
        elif pilihan == '3':
            id_barang = int(input("Masukkan ID barang yang ingin dihapus: "))
            daftar_barang.hapus_barang(id_barang)
            print(f"Barang dengan ID {id_barang} berhasil dihapus.")
        elif pilihan == '4':
            daftar_barang.daftar_semua_barang()
        elif pilihan == '5':
            print("Terima kasih, Sampai Jumpa!")
            break
        else:
            print("Pilihan tidak valid. Silakan masukkan nomor menu yang benar.")

if __name__ == "__main__":
    main()
```

Dengan menggunakan pohon biner pencarian, program ini memungkinkan pengguna untuk dengan mudah menambah, mencari, menghapus, dan menampilkan daftar barang dengan efisien. Implementasi BST memastikan bahwa operasi-operasi tersebut dilakukan dalam waktu logaritmik, menjadikannya pilihan yang baik untuk aplikasi manajemen data.

Pohon Biner Pencarian (BST):
- Pohon biner pencarian adalah struktur data berhirarki di mana setiap simpul memiliki dua anak simpul (kiri dan kanan).
- Dalam BST, setiap simpul memiliki nilai kunci yang unik, dan nilai di simpul kiri lebih kecil daripada nilai di simpul tersebut, sedangkan nilai di simpul kanan lebih besar.
- BST memungkinkan operasi pencarian, penambahan, dan penghapusan dengan kompleksitas waktu rata-rata O(log n), di mana n adalah jumlah elemen dalam pohon.

Kelas TreeNode:
- Kelas TreeNode merepresentasikan simpul dalam BST.
- Setiap simpul memiliki atribut id_barang (kunci), nama, dan harga untuk menyimpan informasi tentang barang, serta left dan --right untuk menunjukkan ke simpul anak kiri dan kanan.

Kelas BinarySearchTreeBarang:
- Kelas BinarySearchTreeBarang mengimplementasikan BST untuk menyimpan dan mengelola data barang.
- Metode tambah_barang digunakan untuk menambahkan barang ke pohon BST, sedangkan metode cari_barang digunakan untuk mencari barang berdasarkan ID-nya.
- Metode hapus_barang digunakan untuk menghapus barang berdasarkan ID-nya dari pohon BST.
- Metode daftar_semua_barang digunakan untuk menampilkan semua barang yang ada dalam pohon BST.

Main Function (main):
- Fungsi main adalah titik masuk utama program.
- Di dalam loop while, pengguna diminta untuk memilih menu untuk menambah, mencari, menghapus, menampilkan semua barang, atau keluar dari program.
- Setiap menu memanggil metode yang sesuai dari kelas BinarySearchTreeBarang untuk menangani operasi yang diminta pengguna.
